```python
import pandas as pd
```

```python
df = pd.read_csv("Data/starbucks.csv",index_col = "Date", parse_dates = True)
```

```python
df.head()
```

```python
df['Close'].plot(figsize = (12,5))
```

```python
## rolling mean 
df.rolling(window = 7).mean()
# from the current row takes last 7 rows including current and finds mean
```

```python
df['Close'].plot(figsize = (12,7))
df.rolling(window = 30).mean()['Close'].plot()
```

```python
## in the bot left there is no info of first 29 days 
```

```python
# with legend
df['Close 30 day mean']  = df['Close'].rolling(window = 30).mean()
df.head()
```

```python
df[['Close','Close 30 day mean']].plot(figsize= (12,5))
```

```python
## explaning
## instead of taking fixed window we want to take values from start to time series to each point in time
df['Close'].expanding().mean().plot(figsize = (12,4))
```
